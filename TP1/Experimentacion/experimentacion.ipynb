Corrimos ej3_perf_test con distintos inputs con una cantidad de números par, estos input respetan siempre que $s_i < t_i$ para todo $1 \leq i \leq n$

### Gráfico con instacias aleatorias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure()

df = pd.read_csv("./runtime_random.csv")
df['cota'] = df['n']

ax1 = sns.lmplot(x='time', y='cota', data=df)
plt.xlabel("tiempo (s)")
plt.ylabel("cota teorica")
```

### Gráfico con instancias aleatorias (escala logarítmica)

```python
df['log_time'] = np.log(df['time'])
df['log_cota'] = np.log(df['cota'])
ax1 = sns.lmplot(x='log_time', y='log_cota', data=df)
plt.xlabel("log_tiempo (s)")
plt.ylabel("log_cota")
```

### Coeficiente de correlación de Pearson

```python
r = np.corrcoef(df['log_time'], df['log_cota'])[0, 1];
print("El coeficiente de correlación de Pearson es: r="+str(r))
```

## Comparación entre instancias ordenadas y desordenadas

Para esta comparación, se crearon instancias que se encuentran ordenadas de forma creciente, con el fin de comparar esta familia de instancias con la anterior (random)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./runtime_ordenado.csv")
df2 = pd.read_csv("./runtime_random.csv")
df['time_random'] = df2['time']

fig, ax = plt.subplots()

ax.scatter(np.log(df['n']),np.log(df['time']),color='red')
ax.plot(np.log(df['n']),np.log(df['time']),color='red', label='ordenado')
ax.scatter(np.log(df['n']),np.log(df['time_random']),color='blue')
ax.plot(np.log(df['n']),np.log(df['time_random']),color='blue', label='random')

plt.xlabel("log_n")
plt.ylabel("log_tiempo (s)")

ax.legend()
ax.grid(True)

plt.show()
```

```python
log_c = np.average(np.log(df['time_random']) - np.log(df['time']))
print("La constante entre ambos conjuntos de datos es c=" + str(np.exp(log_c)))
```
